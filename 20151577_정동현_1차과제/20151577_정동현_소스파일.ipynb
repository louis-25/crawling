```python
# selenium 라이브러리 불러오기
from selenium import webdriver
from datetime import datetime

# 크롬 드라이버를 실행
driver = webdriver.Chrome(r'D:\File\Util\chromedriver_win32\chromedriver.exe') # '' 안에 자신이 설치한 경로

# 네이버 뉴스검색
search_keyword = '코로나' #원하는 키워드 입력하기

# date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

url = f'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={search_keyword}'
driver.get(url)

# html 다운로드
html = driver.page_source
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'html.parser') # html 파싱
```

```python
item = soup.select('#main_pack > div.news.mynews.section._prs_nws > ul > li')

#html파싱이 어떤식으로 되는지 파악하기
for i in range(10):    
    sp_nws = (str)(item[i]['id'])    
    article = soup.select(f'#{sp_nws} > dl > dd.txt_inline')
    print(article[0].text.split(' '))
    article = soup.select(f'#{sp_nws} > dl > dd.txt_inline > span._sp_each_source')
    print(article[0].text.split('언론사')[0])
    article = soup.select(f'#{sp_nws} > dl > dt > a')

    print(article[0]['title'])
    print(article[0]['href'])
    
    print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
```

```python
# 전체 페이지 목록
# paging = driver.find_element_by_class_name('paging')
# 페이지 목록 안에 a태그 추출
# page_1 = paging.find_element_by_id('pagerTagAnchor1')
data = []
page = 0

import time

while(True):
    # 다음페이지 여부 확인
    print('@@@@@@', page+1)
    item = soup.select('#main_pack > div.news.mynews.section._prs_nws > ul > li')      
    
    for i in range(10):    
        sp_nws = (str)(item[i]['id'])
        article = soup.select(f'#{sp_nws} > dl > dd.txt_inline > span._sp_each_source')
        press = article[0].text.split('언론사')[0]
        print("언론사 : ",press)
        article = soup.select(f'#{sp_nws} > dl > dt > a')        
        print("제목 : ",article[0]['title'])
        print("링크 : ",article[0]['href'])
        print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
        data.append([datetime.today().strftime("%Y-%m-%d"), "정동현", search_keyword, press, article[0]['title'], article[0]['href']])
                
    try :        
        if page > 300:
            raise StopIteration
        time.sleep(1)        
        page = page + 1
        if page == 1:
            driver.find_element_by_xpath('//*[@id="main_pack"]/div[1]/div[2]/a[10]').click()        
        else :
            driver.find_element_by_xpath('//*[@id="main_pack"]/div[1]/div[2]/a[11]').click()        
            
    except StopIteration:
        print("다음 페이지 없음 !! \n")
        break
```

```python
# selenium 라이브러리 불러오기
from selenium import webdriver

# 크롬 드라이버를 실행
driver = webdriver.Chrome(r'D:\File\Util\chromedriver_win32\chromedriver.exe') # '' 안에 자신이 설치한 경로

# 네이버 뉴스검색
search_keyword = '코로나' #원하는 키워드 입력하기
url = f'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={search_keyword}'
driver.get(url)

# html 다운로드
html = driver.page_source
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'html.parser') # html 파싱
```

```python
#item = soup.select('_sp_each_title') # 클래스 이름으로 접근

#print(item[0]['title'])
#print(item[0]['href'])

item = soup.select('#main_pack > div.news.mynews.section._prs_nws > ul > li')
# print(item)

for i in range(10):    
    sp_nws = (str)(item[i]['id'])    
    article = soup.select(f'#{sp_nws} > dl > dd.txt_inline')
    print(article[0].text.split(' '))
    article = soup.select(f'#{sp_nws} > dl > dd.txt_inline > span._sp_each_source')
    print(article[0].text.split('언론사')[0])
    article = soup.select(f'#{sp_nws} > dl > dt > a')
#     print(item[i]['id'])
    print(article[0]['title'])
    print(article[0]['href'])
    
    print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
```

```python
# 전체 페이지 목록
# paging = driver.find_element_by_class_name('paging')
# 페이지 목록 안에 a태그 추출
# page_1 = paging.find_element_by_id('pagerTagAnchor1')
data = []
page = 0

import time

while(True):
    # 다음페이지 여부 확인
    print('@@@@@@', page+1)
    item = soup.select('#main_pack > div.news.mynews.section._prs_nws > ul > li')
    # print(item)
    print(item[page]['id']+'###############################################################')
    
    for i in range(10):    
        sp_nws = (str)(item[i]['id'])
        article = soup.select(f'#{sp_nws} > dl > dd.txt_inline > span._sp_each_source')
        print("언론사 : ",article[0].text.split('언론사')[0])
        article = soup.select(f'#{sp_nws} > dl > dt > a')        
        print("제목 : ",article[0]['title'])
        print("링크 : ",article[0]['href'])
        print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
        data.append([search_keyword, article[0]['title'], article[0]['href']])
    page = page + 1
    try :        
        if page > 300:
            raise StopIteration
            
        driver.find_element_by_xpath('//*[@id="main_pack"]/div[1]/div[2]/a[%s]' %page).click()
        time.sleep(1)
    except StopIteration:
        print("다음 페이지 없음 !! \n")
        break
```

```python
import pandas as pd
from konlpy.tag import Kkma
from pandas import Series, DataFrame
 
data = pd.read_csv("20151577_정동현_결과파일.csv",encoding='utf-8',engine='python',usecols=['제목'])
len(data)

kkma = Kkma()
ex_nouns=[]
# data
for i, row in data.iterrows():
    for noun in kkma.nouns(row[0]):
        ex_nouns.append(noun)

ex_nouns
len(ex_nouns)
print(ex_nouns)
```

```python
from re import match

word_count={}

for noun in ex_nouns:
    word_count[noun] = word_count.get(noun,0)+1
    
word_count
```

```python
from collections import Counter

counter = Counter(word_count)

top100 = counter.most_common(100)
top100
```

```python
import pytagcloud
import webbrowser

word_count_list=pytagcloud.make_tags(top100, maxsize=80)

print(word_count_list)

pytagcloud.create_tag_image(word_count_list,
                           'wordcloud.jpg',
                           size=(900,600),
                           fontname='korean',
                           rectangular=False)
webbrowser.open('wordcloud.jpg')
```
